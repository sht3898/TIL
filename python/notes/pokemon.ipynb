```python
import random

class Pikachu:
    
    def __init__(self, name):
        self.name = name
        self.level = 5
        self.hp = self.level * 20 * random.randint(15,20)
        self.exp = 0
        print(f'{self.name}의 HP가 {self.hp}로 설정되었습니다.')
    
    def body_attack(self, enemy):
        print(f'{self.name}이 {enemy.name}을 몸통박치기 공격을 했다....!')
        if random.randint(0, 1):
            enemy.hp -= self.level * 5
            print(f'{enemy.name}의 체력이 {enemy.hp}가 되었다.')
        else:
            print('공격에 실패했다.')
            
    def thousond_volt(self, pokemon):
        print(f'{self.name}이 {enemy.name}을 몸통박치기 공격을 했다....!')
        if random.randint(0, 1):
            enemy.hp -= self.level * 7
            print(f'{enemy.name}의 체력이 {enemy.hp}가 되었다.')
        else:
            print('공격에 실패했다.')
            
    def __repr__(self):
        return f'{self.name}입니다'
        
```

```python
turn = 0
p1_name = input('플레이어 1의 이름을 정해주세요.')
p2_name = input('플레이어 2의 이름을 정해주세요.')
p1 = Pikachu(p1_name)
p2 = Pikachu(p2_name)
print('=============게임이 시작되었습니다!=============')
while True:
    print(f'============={turn}턴=============')
    if turn % 2:
        p1.body_attack(p2)
    else:
        p2.body_attack(p1)
    if p1.hp <= 0 or p2.hp <= 0:
        if p1.hp > p2.hp:
            print(f'{p1.name}이(가) 승리하였습니다.')
            break
        else:
            print(f'{p2.name}이(가) 승리하였습니다.')
            break
    turn += 1
```
